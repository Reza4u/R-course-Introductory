# 2. কেন আমরা R ব্যবহার করি? (Why researchers use R?)

আপনার মনে প্রশ্ন জাগতে পারে: *কেন আমরা Excel বা SPSS এর বদলে R শিখছি?*

Economics বা Public Health-এর গবেষক হিসেবে R আপনাকে কয়েকটি বিশেষ "সুপারপাওয়ার" দেবে:
1. **Reproducibility**: আপনি একটি Script তৈরি করে রাখতে পারেন যা raw data প্রসেস করে, models রান করে এবং স্বয়ংক্রিয়ভাবে রিপোর্ট তৈরি করে। নতুন ডেটা এলে শুধু Script-টি পুনরায় রান করলেই কাজ হয়ে যায়!
2. **Handling Huge Datasets**: লক্ষ লক্ষ rows-এর ডেটা নিয়ে কাজ করতে গেলে Excel প্রায়ই স্লো হয়ে যায় বা ক্র্যাশ করে। কিন্তু R খুব সহজেই বড় ডেটাসেট (Huge Datasets) হ্যান্ডেল করতে পারে।
3. **Publication-Quality Graphics**: একাডেমিক জার্নালে পাবলিশ করার মতো আকর্ষণীয় এবং প্রফেশনাল Data Visualizations তৈরি করার জন্য R হলো স্ট্যান্ডার্ড।


# 3. R এবং RStudio-এর ইন্টারফেস পরিচিতি (Opening R for the First Time)

RStudio ওপেন করলে আপনি সাধারণত চারটি প্যান (panes) দেখতে পাবেন:

* **The Console (নিচে-বামে)**: এটি হলো মূল "ইঞ্জিন"। আপনি এখানে সরাসরি কমান্ড লিখতে পারেন এবং Enter চাপলে সাথে সাথেই রেজাল্ট দেখতে পাবেন। কিন্তু এখান থেকে কোড সেভ করা যায় না।
* **The Editor / Script Window (উপরে-বামে)**: এখানে আপনি আপনার Code লিখে সংরক্ষণ (Save) করে রাখতে পারবেন (যেমন কোনো রেসিপি লিখে রাখা)। আপনি এখান থেকে Code রান করে Console-এ রেজাল্ট পাঠাতে পারবেন। 
* **The Environment (উপরে-ডানে)**: এটি আপনার Memory-তে ঠিক এই মুহূর্তে কী কী Data এবং Variables লোড করা আছে তা দেখায়।
* **Files/Plots/Packages/Help (নিচে-ডানে)**: এই প্যানটি আপনার কম্পিউটারের Files, গ্রাফ (Plots), ইনস্টল করা Packages এবং Help Manual-গুলো দেখতে কাজে লাগে।

# 6. R Packages ইনস্টল, আপডেট এবং লোড করা (Installing, Updating, and Loading R Packages)

R-এ কিছু বিল্ট-ইন ফাংশন থাকে, তবে এর আসল শক্তি লুকিয়ে আছে **Packages**-এর ভেতরে। প্যাকেজ হলো অন্য গবেষকদের লেখা কোডের একটি বান্ডিল, যা আপনি বিনামূল্যে ডাউনলোড করে ব্যবহার করতে পারেন।

* **ইন্সটল করা (Installing):** একবার কোনো প্যাকেজ পিসিতে ডাউনলোড করার জন্য `install.packages("packagename")` ফাংশন ব্যবহার করতে হয়।
* **আপডেট করা (Updating):** আপনার ইনস্টল করা প্যাকেজগুলোর নতুন ভার্সন চেক এবং আপডেট করতে `update.packages()` ব্যবহার করতে হয়।
* **লোড করা (Loading):** কোনো প্যাকেজ ব্যবহার করার আগে সেটিকে আপনার বর্তমান সেশনে চালু করে নিতে হয়। এর জন্য `library(packagename)` ফাংশন ব্যবহার করতে হয় (এটি প্রতিবার RStudio ওপেন করার পর করতে হয়)।

# 7. Help Files এবং Function Documentation ব্যবহার করা (Getting Help)

সবচেয়ে অভিজ্ঞ প্রোগ্রামাররাও ফাংশনের কাজ ভুলে যেতে পারেন! আপনি যদি কোনো ফাংশন সম্পর্কে না জানেন বা সেটি কীভাবে কাজ করে তা বিস্তারিত শিখতে চান, তবে `?` অথবা `help()` ফাংশনটি ব্যবহার করতে পারেন।


```r
# mean() ফাংশনের ডকুমেন্টেশন চেক করতে:
?mean

# অথবা:
help("mean")
```

যদি আপনি নির্দিষ্ট কোনো ফাংশনের নাম ভুলে যান, তবে একটি শব্দ দিয়ে পুরো R প্যাকেজগুলোতে সেটি কোথায় কোথায় আছে তা খুঁজতে পারেন `??` অথবা `help.search()` ব্যবহার করে।


```r
# "mean" শব্দটি সম্পর্কিত ফাংশনগুলো খুঁজতে:
??"mean"

# অথবা:
help.search("mean")
```

# 9. Recap

চলুন আজকের ক্লাসে কী কী শিখলাম তা এক নজরে দেখে নিই:

1.  **কেন R শিখবো?**: Reproducibility, বিশাল ডেটা হ্যান্ডেল করা এবং চমৎকার গ্রাফ বা প্লট তৈরি করতে।
2.  **RStudio-এর বিভিন্ন অংশ**: Console এবং Editor-এর মধ্যকার পার্থক্য।
3.  **Comments**: `#` ব্যবহার করে কোডটি কীভাবে কাজ করে তা লিখে রাখা।
4.  **Working Directory**: `getwd()` এবং `setwd()` কমান্ড দিয়ে ফাইলের অবস্থান চেক করা এবং পরিবর্তন করা।
5.  **Packages**: `install.packages()`-এর মাধ্যমে প্যাকেজ ডাউনলোড করা, `update.packages()` দিয়ে আপডেট করা, এবং `library()` দিয়ে তা সেশনে লোড করা।
6.  **Help**: `?function_name` বা `??character_string` ব্যবহার করে ডকুমেন্টেশন পড়া।
7.  **Saving & Exiting**: `ls()`, `save.image()`, `load()`, এবং `q()` এর মতো কমান্ড দিয়ে কাজ সেভ করা এবং R থেকে বের হওয়া।
