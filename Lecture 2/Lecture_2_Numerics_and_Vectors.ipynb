# 1. R-এ সাধারণ গাণিতিক হিসাব (R for Basic Math)

R-কে একটি শক্তিশালী সায়েন্টিফিক ক্যালকুলেটর হিসেবে ব্যবহার করা যায়। যোগ (+), বিয়োগ (-), গুণ (*), ভাগ (/) এবং পাওয়ার (^) এর মতো সাধারণ কাজগুলো আমরা সরাসরি Console-এ করতে পারি।


```r
# যোগ, বিয়োগ, গুণ এবং ভাগ
10 + 5
14 / 6
(10 * 2) - 4

# পাওয়ার বা সূচক (Exponent)
3 ^ 2
```

গবেষণায় ডেটা ট্রান্সফর্ম করার জন্য প্রায়ই লগারিদম এবং এক্সপোনেনশিয়াল ফাংশনের প্রয়োজন হয়।


```r
# বর্গমূল (Square Root)
sqrt(9)

# লগারিদম (Natural Log)
log(20.08554)

# এক্সপোনেনশিয়াল (Exponential)
exp(3)
```

# 2. ডেটা সংরক্ষণ বা Assign করা (Assigning Objects)

```r
# একটি অবজেক্টে ডেটা স্টোর করা
monthly_income <- 25000 / 12

# সেই অবজেক্টটিকে দেখতে চাইলে তার নাম লিখুন
monthly_income

# একটি অবজেক্টকে অন্য অবজেক্টের সাথে হিসাব করা
yearly_income <- monthly_income * 12
yearly_income
```

# 3. ভেক্টর: একসাথে অনেকগুলো ডেটা (Vectors)

একটি ভেক্টর হলো একই ধরনের অনেকগুলো ডেটার সংগ্রহ। যেমন ধরুন, গত এক সপ্তাহের প্রতিদিনের রোগীর সংখ্যা। ভেক্টর তৈরি করতে `c()` ফাংশনটি (যার অর্থ Combine বা Concatenate) ব্যবহার করা হয়।


```r
# এক সপ্তাহের রোগীর সংখ্যার একটি ভেক্টর তৈরি
patients_this_week <- c(45, 50, 42, 60, 55, 30, 48)
patients_this_week
```

---
# 4. Sequences, Repetition, Sorting এবং Lengths

ভেক্টর তৈরি এবং সাজানোর জন্য কিছু কার্যকরী ফাংশন:

*   **Sequence (`:` বা `seq()`):** সিরিয়ালি সংখ্যা তৈরি করা।
*   **Repetition (`rep()`):** একই সংখ্যা বা প্যাটার্ন বারবার লেখা।
*   **Sorting (`sort()`):** ছোট থেকে বড় বা বড় থেকে ছোট ক্রমানুসারে সাজানো।
*   **Length (`length()`):** ভেক্টরে মোট কয়টি ডেটা বা এন্ট্রি আছে তা বের করা।


```r
# ১ থেকে ১০ পর্যন্ত সংখ্যা তৈরি (Sequence)
1:10

# জোড় সংখ্যার Sequence তৈরি করা (2 করে বৃদ্ধি)
seq(from = 2, to = 20, by = 2)

# একটি নির্দিষ্ট সংখ্যাকে ৫ বার রিপিট করা
rep(x = 10, times = 5)

# রোগীদের সংখ্যা ছোট থেকে বড় অনুসারে সাজানো
sort(patients_this_week, decreasing = FALSE)

# ভেক্টরে মোট কয়টি এন্ট্রি আছে?
length(patients_this_week)
```

# 5. সাবসেটিং: ভেক্টর থেকে নির্দিষ্ট ডেটা বের করা (Subsetting and Element Extraction)

### Base R স্টাইল (`[ ]`)
ভেক্টর থেকে কোনো নির্দিষ্ট পজিশন বা ইনডেক্স-এর ডেটা বের করতে আমরা থার্ড ব্র্যাকেট `[ ]` ব্যবহার করি।


```r
patients_this_week

# সপ্তাহের প্রথম দিনের রোগীর সংখ্যা
patients_this_week[1]

# তৃতীয় দিন থেকে পঞ্চম দিনের রোগীর সংখ্যা
patients_this_week[3:5]

# দ্বিতীয় দিনের ডেটা বাদে বাকি সব দেখতে চাইলে নেগেটিভ ইনডেক্স ব্যবহার করতে হয়
patients_this_week[-2]
```

### Tidyverse স্টাইল (Tidyverse Thinking)
যেহেতু আমরা কোর্সে মুলত Tidyverse ব্যবহার শিখবো, তাই আমরা দেখবো কিভাবে ফিল্টার (Filter) করে ডেটা সাবসেট করা যায়। সাধারণ ভেক্টরে ফিল্টার কাজ করলেও, এটি সাধারণত ডাটাফ্রেমের (Dataframe) ক্ষেত্রে বেশি ব্যবহৃত হয়। তবে আমরা এখানে Tidyverse-এর কনসেপ্ট প্রয়োগ করতে পারি!


```r
library(tidyverse)

# Tidyverse-এ শর্ত সাপেক্ষে ভেক্টর থেকে সাবসেট করার একটি পদ্ধতি (Indexing with Logic)
# কোন দিনগুলোতে রোগীর সংখ্যা ৫০ এর বেশি ছিল?
patients_this_week[patients_this_week > 50]

# Tidyverse -এ সাধারণত আমরা ডাটাফ্রেম বা tibble ব্যবহার করি `filter()` করার জন্য:
patient_df <- tibble(
  day = 1:7,
  patients = patients_this_week
)

# Tidyverse এর filter() ব্যবহার করে ৫০ এর চেয়ে বেশি রোগীর দিনগুলো বের করা
patient_df %>% 
  filter(patients > 50)
```

# 6. ভেক্টর-ওরিয়েন্টেড হিসাব (Vector-Oriented Behavior)

R-এর সবচেয়ে বড় সুবিধা হলো এটি পুরো ভেক্টরের ওপর একসাথে কাজ করতে পারে। একে **Element-wise Math** বলা হয়। আপনাকে লুপ (Loop) লিখে প্রতিটি ডেটার ওপর আলাদাভাবে গাণিতিক কাজ করতে হবে না।


```r
expenses_in_taka <- c(500, 1000, 1500)

# ধরুন, ஒவ்வொரு খরচের সাথে আপনি ১০০ টাকা যোগ করতে চান
expenses_in_taka + 100

# এই সপ্তাহের মোট রোগী কতজন?
sum(patients_this_week)

# ಗುಣফল (Product)
prod(c(2, 3, 4))
```

# 7. Recap

আজকের ক্লাসে আমরা যা শিখলাম:
1.  **গাণিতিক কাজ**: `+`, `-`, `seq()`, `sqrt()`, `exp()`, `log()`
2.  **Assignment**: `<-` দিয়ে ডেটা সেভ করা।
3.  **ভেক্টর তৈরি**: `c()` দিয়ে অনেকগুলো ডেটা একসাথে রাখা।
4.  **সাবসেটিং (Subsetting)**: `[ ]` দিয়ে ইনডেক্স অনুযায়ী ডেটা বের করা এবং Tidyverse-এ Logic বা `filter()` ব্যবহার করা।
5.  **ভেক্টর হিসাব (Vectorized operations)**: একসাথে পুরো ভেক্টরের ওপর গাণিতিক কাজ করা এবং `sum()`, `prod()` ইত্যাদি ব্যবহার করা।
